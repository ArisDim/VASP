```python
#### Written by Aris Dimou 
def Create_VASP(Structure, rep, i_dw, file_n, boolean, Relax_at = ['T', 'T', 'T']):
    
    ### Variables
    
    atom_id, mol_tag, atom_t = 1, 1, 1
    bond_id, bond_type, uc = 1, 1, 5
    
    a, b, c = 3.94149, 3.94149, 3.98740
    a, b, c = 3.94568, 3.94568, 3.98252   ### LDA lattice
    
#     a, b, c = 3.957959,  3.957959, 4.099704
    
    prec = 5
    
    
    ### Lambda functions 
    
    l_coord = (lambda x: f'{x:5f}') ## turns coordinates into strings   
    
    ### Results
    
    A_line, B_line, O_line = [], [], []
    
    x, y, z = np.array([]), np.array([]), np.array([])       # Atom coord

    ##########################################################################
    ##################            Masses            ##########################
    ##########################################################################

    rep_x, rep_y, rep_z = rep[0], rep[1], rep[2]
    NofA = uc*rep_x*rep_y*rep_z               # Number of atoms
    N_cells = rep_x*rep_y*rep_z 
    
    
    at_id =[1, 2, 3, 4, 5, 6, 5, 6, 5, 6]

    if Structure == 'Cubic':
        a, b, c = 3.96769, 3.96769, 3.96769
        x = np.array((0.0, 0.5, 0.0, 0.5, 0.5)) #all x Ba, Ti, O1, O2, O3
        y = np.array((0.0, 0.5, 0.5, 0.0, 0.5)) #all y Ba, Ti, O1, O2, O3
        z = np.array((0.0, 0.5, 0.5, 0.5, 0.0)) #all z Ba, Ti, O1, O2, O3

    elif Structure == 'Tetragonal':

        x = np.array([0.0, 0.5, 0.0, 0.5, 0.5])
        y = np.array((0.0, 0.5, 0.5, 0.0, 0.5))

    #     z = np.array([0.000000, -0.000344, 0.517587, 0.501805,\
    #                       0.485796, 0.486726, 0.485796, 0.486726\
    #                       , -0.040118,-0.069991])*c

        z = np.array([-0.00239, 0.50851, 0.48643, 0.48643, -0.01982])
        z_ = np.array([0.00239, 0.49149, 0.51357, 0.51357, 0.01982])
        
    else:
        print('Error')

    ##########################
    ###     Box Dimensions number of bonds etc
    ##########################
    with open(file_n, 'w') as fw:
        
        fw.write('#Created by Dimou A.\n')
        fw.write('1.00000\n' )
        fw.write('%f 0.00000 0.00000\n'%(round(rep_x*a, 5)))
        fw.write('0.00000 %f 0.00000\n'%(round(rep_y*b, 5)))
        fw.write('0.00000 0.00000 %f\n'%(round(rep_z*c, 5)))
        
        fw.write('Ba Ti O\n')
        fw.write('%i %i %i\n'%(N_cells, N_cells, N_cells*3))
        fw.write('Direct\n')
        
    old_id, atom_id, mol_id =1, 1, 1                                        


    c1, c2 = 0, 0
    for repz in range(0,rep_z):
        for repy in range(0, rep_y):
            for repx in range(0, rep_x):
                _i = 0

                for _id in range(old_id, old_id+uc): #str(format(x[0]+rep_x*a, '.7f'))
                    

                    t_line = []

                    if boolean == True and Structure == 'Tetragonal':

                        if repx+1 <= i_dw:


                            t_xyz = [(x[_i]+repx)/(rep_x), (y[_i]+repy)/(rep_y),
                                     (z[_i]+repz)/(rep_z)]      

                            t_line = list(map(l_coord, t_xyz)) + Relax_at


                            print(t_xyz)
                            
                            c1 += 1


                        else:

                            
                            t_xyz = [(x[_i]+repx)/(rep_x), (y[_i]+repy)/(rep_y),
                                     (z_[_i]+repz)/(rep_z)]      
                                
                            print(t_xyz)

                            t_line = list(map(l_coord, t_xyz)) + Relax_at
                            
                            c2 += 1
                            



                    elif boolean == False or Structure == 'Cubic':


                            t_xyz = [(x[_i]+repx)/(rep_x), (y[_i]+repy)/(rep_y),
                                     (z[_i]+repz)/(rep_z)]                            

                            print(t_xyz, _i)

                            t_line = list(map(l_coord, t_xyz)) + Relax_at
                            #x[_i]+repx*a

                            c1 += 1
                            c2 += 1    
                    
                    #print(t_line)
                    
                    _i += 1

                    if _i == 1:
                        A_line.append(t_line)
                    elif _i == 2:
                        B_line.append(t_line)
                    elif _i > 2:
                        O_line.append(t_line)
                    
                
                old_id += uc
                
    with open(file_n, 'a') as fa:
        for l in A_line:
            line = ' '.join(l)+'\n'
            fa.write(line)
        for l in B_line:
            line = ' '.join(l) +'\n'
            fa.write(line)
        for l in O_line:
            line = ' '.join(l) +'\n'
            fa.write(line)    
            
```
